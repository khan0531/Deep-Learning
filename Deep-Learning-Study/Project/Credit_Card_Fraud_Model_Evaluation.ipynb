### 2.2.3 General Overview

#### **1. F1 Score**
- **의미**:
  - F1 Score는 **정밀도(Precision)**와 **재현율(Recall)**의 조화평균으로, 두 메트릭의 균형을 평가합니다.
  - **불균형 데이터**에서 적합하며, **정확도(Accuracy)**가 높은 가짜 신호를 줄이는 데 유용합니다.
  - 공식: $F1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$

- **언제 중요한가**:
  - **Positive 클래스**(예: 사기 탐지, 질병 진단)가 매우 드물거나, 잘못된 Negative 예측보다 Positive 예측이 더 중요한 경우.
  - Precision(정밀도)과 Recall(재현율) 사이의 균형이 중요할 때.

- **예시**:
  - 사기 탐지 모델에서 사기를 놓치는 것(False Negative)과 잘못된 알림(False Positive)을 최소화하고 싶을 때.

---

#### **2. ROC AUC Score**
- **의미**:
  - ROC AUC(Receiver Operating Characteristic Area Under Curve)는 분류 모델이 **모든 임계값(threshold)**에서 Positive 클래스와 Negative 클래스를 얼마나 잘 구분하는지 평가합니다.
  - AUC 값은 **0.5~1.0** 사이로, 1.0에 가까울수록 좋습니다.

- **언제 중요한가**:
  - **전체적인 분류 성능**을 평가하고 싶을 때.
  - Positive 클래스와 Negative 클래스의 **전반적인 구분 능력**을 확인하고 싶을 때.
  - **클래스 간의 불균형이 심하지 않은 경우**에도 효과적입니다.

- **예시**:
  - 의료 분야에서 환자가 특정 질병에 걸렸을 가능성을 측정할 때, 모델의 **모든 임계값에서 구분 능력**을 평가.

---

#### **3. Precision-Recall AUC**
- **의미**:
  - Precision-Recall AUC는 다양한 **재현율(Recall)**에서 모델의 **정밀도(Precision)**를 평가합니다.
  - 특히 **클래스 불균형 데이터**에 매우 유용하며, Positive 클래스를 올바르게 예측하는 성능을 강조합니다.

- **언제 중요한가**:
  - Positive 클래스(예: 사기, 희귀 질환 등)가 **아주 적은 경우**.
  - False Negative(예: 실제 사기를 놓치는 것)가 더 큰 영향을 미칠 때.
  - ROC AUC가 Positive 클래스에 대해 **너무 낙관적**일 때 사용.

- **예시**:
  - 사기 탐지에서 **사기를 최대한 탐지(Recall)**하면서도 **잘못된 경고(Precision)**를 줄이고 싶을 때.

---

#### **어떤 상황에서 어떤 메트릭을 볼 것인가?**

1. **클래스 불균형이 심한 경우**:
   - 주로 Positive 클래스(예: 사기, 질병) 탐지가 중요한 경우:
     - **F1 Score**: Precision과 Recall의 균형을 확인.
     - **Precision-Recall AUC**: 다양한 Recall에서의 Precision 성능을 확인.

2. **전반적인 분류 성능 평가**:
   - Positive와 Negative 클래스 비율이 상대적으로 균형 잡혀 있을 때:
     - **ROC AUC Score**: 전체 구분 능력을 평가.

3. **정확도가 중요한 경우**:
   - 잘못된 Positive 예측(False Positive)을 최소화해야 할 때:
     - **Precision**이 중요한 경우 F1 Score 또는 Precision-Recall AUC를 활용.

4. **재현율이 중요한 경우**:
   - False Negative가 치명적인 결과를 초래하는 경우(예: 질병 진단, 사기 탐지):
     - **F1 Score** 또는 **Precision-Recall AUC**에 집중.

## 샘플링 전략

### **샘플링 방법**
클래스 불균형 문제를 해결하기 위해 다음과 같은 샘플링 클래스를 정의합니다:

1. **OriginalSampler**
   - **역할**: 원본 데이터를 수정 없이 사용합니다.
   - **특징**: 클래스 불균형이 있는 그대로 반영되므로 모델 성능의 베이스라인 비교에 사용됩니다.

2. **OverSampler**
   - **역할**: 오버샘플링을 적용하여 데이터 균형을 맞춥니다.
   - **특징**:
     - 주요 기법: **SMOTE** (Synthetic Minority Over-sampling Technique).
     - 적은 클래스(소수 클래스) 데이터를 합성하여 샘플 수를 증가시킵니다.

3. **UnderSampler**
   - **역할**: 언더샘플링을 적용하여 데이터 균형을 맞춥니다.
   - **특징**:
     - 다수 클래스 데이터를 줄여 소수 클래스와 균형을 맞춥니다.
     - 데이터 손실로 인해 정보가 부족할 수 있으므로 소규모 데이터셋에 적합합니다.

4. **SMOTETomekSampler**
   - **역할**: SMOTE와 Tomek Links를 결합하여 오버샘플링과 언더샘플링을 동시에 수행합니다.
   - **특징**:
     - SMOTE로 소수 클래스 데이터를 합성한 뒤, Tomek Links를 사용해 중복되거나 애매한 샘플을 제거합니다.
     - 클래스 간의 경계를 더 명확히 하여 모델 학습 효율을 높입니다.

## 3.1. Base Model

```python
class BaseModelTrainer:
    def __init__(self, x_train, y_train, sampling_strategy=None):
        self.x_train = x_train
        self.y_train = y_train
        self.sampling_strategy = sampling_strategy
        self.sampled_data = None
        self.sampled_labels = None

        if sampling_strategy == 'smote':
            smote = SMOTE(sampling_strategy='minority', random_state=42)
            self.sampled_data, self.sampled_labels = smote.fit_resample(x_train, y_train)
        elif sampling_strategy == 'smotetomek':
            smote_tomek = SMOTETomek(sampling_strategy='auto', random_state=42)
            self.sampled_data, self.sampled_labels = smote_tomek.fit_resample(x_train[:20000], y_train[:20000]) # 코랩 메모리 부족으로 데이터 수를 2만개로 한정
        elif sampling_strategy == 'undersample':
            rus = RandomUnderSampler(random_state=42)
            self.sampled_data, self.sampled_labels = rus.fit_resample(x_train, y_train)
        else:
            self.sampled_data, self.sampled_labels = x_train, y_train

    def train_model(self):
        raise NotImplementedError

    def evaluate_model(self, model, x_test, y_test):
        y_pred = model.predict(x_test)

        if hasattr(model, 'predict_proba'):
            y_pred = (y_pred > 0.5).astype(int)

        if isinstance(model, Sequential):  # Check if the model is a neural network
            y_pred = (y_pred >= 0.5).astype(int)

        f1 = f1_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_pred)
        precision, recall, _ = precision_recall_curve(y_test, y_pred)
        pr_auc = auc(recall, precision)
        print(classification_report(y_test, y_pred, target_names=["Not Fraud", "Fraud"]))
        print(f"F1 Score: {f1}")
        print(f"Precision-Recall AUC: {pr_auc}\n")
        return {
            'model': model,
            'f1_score': f1,
            'pr_auc': pr_auc
        }
```

## 3.2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Logistic Regression Trainer class
class LogisticRegressionTrainer(BaseModelTrainer):
    def train_model(self):
        logistic_params = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2'], 'solver': ['liblinear']}
        logistic_model = LogisticRegression(max_iter=1000)
        skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        logistic_grid = GridSearchCV(logistic_model, logistic_params, scoring='f1', cv=skf, verbose=3)
        logistic_grid.fit(self.sampled_data, self.sampled_labels)
        return logistic_grid.best_estimator_
```

## 3.6. Neural Network

###**Focal Loss**
현재 데이터셋은 클래스 불균형이 심한 상태로, Cross Entropy Loss를 사용할 경우 정상 거래에 지나치게 치우친 결과가 도출될 가능성이 있습니다. 이를 보완하기 위해, Focal Loss를 적용하여 사기 거래에 더 집중하고, 균형 잡힌 결과를 도출하고자 합니다.


\begin{equation*}
FL(p_t) = -\alpha_t (1 - p_t)^\gamma \log(p_t)⋅CrossEntropy(y_{true} , y_{pred})
\end{equation*}


- $𝑦_{true}$ : 실제 정답 레이블
- $y_{pred}$ : 모델의 예측 확률
- Cross Entropy Loss: $CrossEntropy(y_{true},y_{pred})$
- $p_t$ : 모델의 올바른 클래스에 대한 확률
- $(1 - p_t)^\gamma$ : 사기 거래(소수 클래스) 강조
- $𝛼$ : 사기 거래(소수 클래스) 가중치

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint

# Deep Learning Trainer class
class DeepLearningTrainer(BaseModelTrainer):
    def train_model(self):
        class FocalLoss(tf.keras.losses.Loss):
            def __init__(self, gamma=2., alpha=0.25, **kwargs):
                self.gamma = gamma
                self.alpha = alpha
                super(FocalLoss, self).__init__(**kwargs)

            def call(self, y_true, y_pred):
                y_true = tf.cast(tf.reshape(y_true, (-1, 1)), tf.float32)
                y_pred = tf.cast(y_pred, tf.float32)
                cross_entropy = tf.keras.losses.binary_crossentropy(y_true, y_pred, from_logits=False)
                pt = tf.exp(-cross_entropy)
                focal_loss = self.alpha * tf.pow(1. - pt, self.gamma) * cross_entropy
                return focal_loss

        nn_model = Sequential([
            Input(shape=(self.sampled_data.shape[1],)),
            Dense(128, activation='relu'),
            Dropout(0.5),
            BatchNormalization(),
            Dense(64, activation='relu'),
            Dropout(0.5),
            BatchNormalization(),
            Dense(1, activation='sigmoid')
        ])

        nn_model.compile(optimizer=Adam(),
                         loss=FocalLoss(),
                         metrics=['Precision', 'Recall', 'AUC'])

        reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-5)
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
        model_checkpoint = ModelCheckpoint('best_nn_model.keras', save_best_only=True, monitor='val_loss')

        nn_model.fit(self.sampled_data, self.sampled_labels,
                     validation_split=0.2,
                     epochs=20,
                     batch_size=32,
                     callbacks=[reduce_lr, early_stopping, model_checkpoint],
                     verbose=1)

        return nn_model
```

## 4.2. SMOTE (Synthetic Minority Over-sampling Technique) Data set

- **SMOTE**는 **소수 클래스**의 데이터를 인위적으로 생성하여 클래스 불균형 문제를 해결하는 **오버샘플링 기법**입니다.

### 작동 방식
1. **소수 클래스 샘플 선택**:
   - 소수 클래스 데이터 중 하나를 임의로 선택합니다.
2. **k-최근접 이웃 탐색**:
   - 선택된 소수 클래스 샘플의 k개의 가장 가까운 이웃을 찾습니다.
3. **새로운 데이터 생성**:
   - 선택된 샘플과 이웃 샘플 간의 선형 간격 내에서 임의의 점을 생성하여 새로운 데이터를 만듭니다.

### 주요 특징

## 4.4. SMOTE-Tomek Data Set

### 개요
**SMOTE-Tomek**은 **SMOTE**(Synthetic Minority Over-sampling Technique)와 **Tomek Links**를 결합한 **혼합 샘플링 기법**으로, 클래스 불균형 문제를 해결하기 위해 **오버샘플링**과 **언더샘플링**을 동시에 수행합니다.


### 작동 방식
1. **SMOTE 적용**:
   - 소수 클래스의 데이터를 기반으로 새로운 합성 데이터를 생성하여 오버샘플링을 수행합니다.
2. **Tomek Links 적용**:
   - 서로 다른 클래스에 속하는 두 샘플이 서로의 최근접 이웃인 경우를 **Tomek Link**라고 정의합니다.
   - 이러한 Tomek Link를 찾아 다수 클래스의 샘플을 제거하여 언더샘플링을 수행합니다.


### 주요 특징
- **오버샘플링과 언더샘플링의 결합**: SMOTE로 소수 클래스의 데이터를 증가시키고, Tomek Links로 다수 클래스의 경계에 있는 샘플을 제거하여 데이터의 균형을 맞춥니다.
- **경계 명확화**: Tomek Links를 통해 클래스 간 경계를 명확히 하여 모델의 분류 성능을 향상시킵니다.


### 장점
- **노이즈 감소**: Tomek Links를 통해 경계에 위치한 노이즈 데이터를 제거하여 모델의 일반화 성능을 높입니다.
- **효율적인 데이터 균형화**: 오버샘플링과 언더샘플링을 동시에 수행하여 데이터의 균형을 효과적으로 맞춥니다.


### 단점
- **계산 비용 증가**: 두 가지 샘플링 기법을 연속적으로 적용하므로 계산 비용이 증가할 수 있습니다.
- **복잡성 증가**: 파라미터 설정과 적용 순서에 따라 결과가 달라질 수 있어 주의가 필요합니다.
